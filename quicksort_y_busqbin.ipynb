```python
'''
Este programa tiene como objetivo hacer la comparacion de dos metodos
de ordenamiento: QuickSort y Búsqueda Binaria. El primero tiene una complejidad
de O(n log n), mientras que el segundo tiene una complejidad de O(log n).
'''
import random
import time
import matplotlib.pyplot as plt
import seaborn as sns

#Función para generar números aleatorios
def Generador(N):
    start_time = time.time()
    lista = [random.randint(1, 5000) for _ in range(N)]
    end_time = time.time()
    generador_time = end_time - start_time
    return lista, generador_time

#Función de ordenamiento QuickSort
def Quicksort(arr):
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[len(arr) // 2]
        left = [x for x in arr if x < pivot]
        middle = [x for x in arr if x == pivot]
        right = [x for x in arr if x > pivot]
        return Quicksort(left) + middle + Quicksort(right)

#Función de búsqueda binaria
def Búsqueda_Binaria(arr, x):
    izquierda, derecha = 0, len(arr) - 1
    while izquierda <= derecha:
        medio = (izquierda + derecha) // 2
        if arr[medio] == x:
            return medio
        elif arr[medio] < x:
            izquierda = medio + 1
        else:
            derecha = medio - 1
    return -1

#Función para analizar el rendimiento y graficar
def Analizar_y_Graficar():
    tamaños = [100, 500, 1000, 2000, 3000, 4000, 5000]
    tiempos_quick = []
    tiempos_busqueda = []

    for N in tamaños:
        lista, _ = Generador(N)

        #Ordenar la lista con QuickSort y medir el tiempo
        start_time = time.time()
        lista_ordenada = Quicksort(lista)
        end_time = time.time()
        quicksort_time = end_time - start_time

        #Buscar un elemento en la lista ordenada y medir el tiempo
        elemento_a_buscar = random.choice(lista_ordenada)
        start_time = time.time()
        Búsqueda_Binaria(lista_ordenada, elemento_a_buscar)
        end_time = time.time()
        busqueda_time = end_time - start_time

        tiempos_quick.append(quicksort_time)
        tiempos_busqueda.append(busqueda_time)

    #Graficar QuickSort
    sns.lineplot(x=tamaños, y=tiempos_quick, marker='o', linestyle='--', label='QuickSort', color='lightseagreen')
    plt.xlabel('Tamaño de la Lista')
    plt.ylabel('Tiempo en segundos')
    plt.title('QuickSort')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Graficar Búsqueda Binaria
    sns.lineplot(x=tamaños, y=tiempos_busqueda, marker='o', linestyle='-', label='Búsqueda Binaria', color='rebeccapurple')
    plt.xlabel('Tamaño de la Lista')
    plt.ylabel('Tiempo en segundos')
    plt.title('Búsqueda Binaria')
    plt.legend()
    plt.grid(True)
    plt.show()

# Llamar a la función para analizar y graficar
Analizar_y_Graficar()
```
